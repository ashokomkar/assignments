```python
sale_rate=pd.cut(df['Sales'],bins=[-1,7.490,16.270],labels=['low sales','high sales'])
sale_rate
```

```python
df['Sales']=sale_rate
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

df['Sales']=1-LE.fit_transform(df['Sales'])
df['ShelveLoc']=LE.fit_transform(df['ShelveLoc'])
df['Urban']=LE.fit_transform(df['Urban'])
df['US']=LE.fit_transform(df['US'])
```

```python
X=df.drop(['Sales'],axis=1)
Y=df['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(max_depth=3,class_weight={0:0.6,1:0.4})
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
import seaborn as sns
sns.set_style(style='darkgrid')
sns.pairplot(dt)
```

```python
# create the dummies
dt=pd.get_dummies(dt,columns=['Undergrad','Marital.Status','Urban'],drop_first=True)
dt['Taxable']=pd.cut(dt['Taxable.Income'],bins=[10002,30000,99620],labels=['Risky','Good'])
```

```python
#After creation of new col. TaxInc also made its dummies var concating right side of df
dt = pd.get_dummies(dt,columns = ["Taxable"],drop_first=True)
```

```python
# normalising the data
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
# Normalized data frame (considering the numerical part of data)
dt_norm = norm_func(dt.iloc[:,1:])
dt_norm.tail()
```

```python
# split variables
x=dt_norm.drop(['Taxable_Good'],axis=1)
y=dt_norm['Taxable_Good']
```

```python
# traing and testing data set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# model fitting
model2=RandomForestClassifier(max_depth=3)
model2.fit(x_train,y_train)
```

```python
# training accuracy
model2.score(x_train,y_train)
```

```python
# testing accuracy
model2.score(x_test,y_test)
```
