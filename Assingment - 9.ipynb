# Read the data

```python
import pandas as pd  
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
```

```python
WIN = pd.read_csv("wine (1).csv")
WIN.head()
```

```python
WIN.tail()
```

```python
WIN.append
```

```python

# Considering only numerical data 
WIN.data = WIN.iloc[:,1:]
WIN.data.head()
# Converting into numpy array
WIN1 = WIN.data.values
WIN1
```

```python
WIN_normal = scale(WIN1)
WIN_normal
```

```python
pca = PCA(n_components=5) 
pca_values = pca.fit_transform(WIN_normal)
pca_values
```

```python
#loadings or weights
pca.components_
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
finalWIN = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), WIN[["Alcohol","Malic"]]], axis = 1)
finalWIN
```

# Hierarchical Clustering

```python
def norm_func(i):
    x = (i - i.min()/i.max()-i.min())
    return (x)
```

```python
WIN_norm = norm_func(WIN.iloc[:,1:])
WIN_norm
```

```python
from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()
data = pd.DataFrame(MM.fit_transform(WIN.iloc[:,1:]))
data 
```

# for creating dendrogram

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sh 
```

# converting into numpy array format

```python
z = linkage(WIN_norm, method="ward",metric="euclidean")
plt.figure(figsize=(15,8))
plt.xlabel("Index")
plt.ylabel("Distence")
plt.title("Hierarchical Clustering Dendrogram")
sh.dendrogram(z)
plt.show()
```

# creating a new column and assigning it to new column

```python
from sklearn.cluster import AgglomerativeClustering

h_complete = AgglomerativeClustering(n_clusters=4, linkage='ward',affinity = "euclidean").fit(WIN_norm) 
 
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
WIN['clust']=cluster_labels
WIN
```

```python
WIN.iloc[:,1:].groupby(WIN.clust).mean()
```

# inertia is another name for WCSS

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.cluster import KMeans
fig = plt.figure(figsize=(15,8))
WCSS = []
for i in range(1, 11):
    clf=KMeans(n_clusters=i)
    clf.fit(WIN_norm)
    WCSS.append(clf.inertia_)
    
```

```python
plt.plot(range(1, 11),WCSS)
plt.title("The Elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```
